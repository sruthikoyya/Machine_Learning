## Support Vector Classifier Implimentation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
## Predictions
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(f"Classification Report: {classification_report(y_test,y_pred)}")
```

```python
print(f"Accuracy is :{accuracy_score(y_test,y_pred)}")
```

```python
print(f"Confusion matrix :{confusion_matrix(y_test,y_pred)}")
```

Sometimes the data is not linearly separable in this situation we can no tuse the linear kernal it gives the bad accuracy insted we use some specific kernals like rbf ,etc.....

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred=rbf.predict(X_test)
```

```python
print(f"After using the rbf kernel the Classification Report is \n:{classification_report(y_test,y_pred)}")
```

```python
print(f"After using the rbf kernel the Accuracy is \n:{accuracy_score(y_test,y_pred)}")
```

```python
print(f"After using the rbf kernel the Confusion matrix is :{confusion_matrix(y_test,y_pred)}")
```

By uisng Polynomial Kernel

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
y_pred=polynomial.predict(X_test)
print(f"After using the Polynomial kernel the Classification Report is \n:{classification_report(y_test,y_pred)}")
print(f"\n After using the polynomial kernel the Accuracy is \n:{accuracy_score(y_test,y_pred)}")
print(f"\n After using the polynomial kernel the Confusion Matrix is {confusion_matrix(y_test,y_pred)}")
```

```python
sigmoid_kernel=SVC(kernel='sigmoid')
sigmoid_kernel.fit(X_train,y_train)
y_pred=sigmoid_kernel.predict(X_test)
print(f"After using the sigmoid_kernel the Classification Report is \n:{classification_report(y_test,y_pred)}")
print(f"\n After using the sigmoid_kernel the Accuracy is \n:{accuracy_score(y_test,y_pred)}")
print(f"\n After using the sigmoid_kernel the Confusion Matrix is {confusion_matrix(y_test,y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV

#defining parameter range
param_grid={'C':[0.1,1,10,100,1000],
           'gamma':[1,0.1,0.01,0.001,0.0001],
           'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
print(f" Classification Report is \n:{classification_report(y_test,y_pred)}")
print(f"\nAccuracy is \n:{accuracy_score(y_test,y_pred)}")
print(f"\nConfusion Matrix is {confusion_matrix(y_test,y_pred)}")
```
